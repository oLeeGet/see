### 資料結構(Data_Structure)

#### 數值
`int, float, bool`

* int = interger(整數) e.g. 0, 123, 99999, -10
* float = float(浮點數) e.g. 0.1, 12.3, 0.99999, -99.1
* bool = boolean(布林值) e.g. True, False

#### 字串
`str, chr`
* str = string(字串) e.g. "ABC", "123", "你好"
* chr = char(字元) e.g. "A", "a", "66"->"B" (通常用於 ASCII code 轉換)

#### 容器
`list, dict, typle, set`
* list = list(有順序, 可動) e.g. [1, "abc", True, 0.5, [1,2,3]]
* dict = dictionary(字典) e.g. {"name":"Alan Fung", "sales_date":"20220529"}
* typle = tuple(有順序, 不可動) e.g. (1, 0.5, [1,2,3], False) //前後以 () 包住, 以逗號","分隔每筆資料, 元組資料無法變更
* set = (集合) e.g. {3,4,5,6,7}

```python
#列表(List)
option1 = ['A大樓','B公寓','C農場','D別墅']
print(type(option1))
```

```python
#選取
print('[Demo 1]')
print(option1[0])
print(option1[1])
print(option1[2])
print(option1[3])
print('-' * 20)
print('[Demo 2]')
print(option1[0:3])
print('-' * 20)
print('[Demo 3]')
x = [[3,4,5,6,7],[7,8,9,10,11]]
print(x[0]) #[3,4,5,6,7]
print(x[0][2]) #[3,4,5,6,7] 中的 5
print(x[1][4]) #[7,8,9,10,11] 中的 11
print('-' * 20)
print('[Demo 4]')
x="Helllo"
print(x[1])
print(x[1:])
print(x[1:4])
print(x[:3])
```

```python
#增加(方法1: 在最後) - append
option1 = ['A大樓','B公寓','C農場','D別墅'] #temp reset
print('變身前:', option1)
option1.append('E廁所')
print('變身後:', option1)
```

```python
#增加(方法2: 在指定位置) - insert
option1 = ['A大樓','B公寓','C農場','D別墅'] #temp reset
print('變身前:', option1)
option1.insert(2,'E廁所')
print('變身後:', option1)
```

```python
#增加(方法3)
print('變身前:', option1)
option1 = option1 + ['E汽車']
print('變身後:', option1)
```

```python
#刪除 - remove
option1 = ['A大樓','B公寓','C農場','D別墅'] #temp reset
print('變身前:', option1)
option1.remove('C農場')
print('變身後:', option1)
```

```python
#合併(方法1) - extend
list1 = ['A大樓','B公寓','C農場','D別墅'] #temp reset
list2 = [100,200,300,400]

print('變身前:', list1, list2)
list1.extend(list2)
print('變身後:', list1)

list1 = ['A大樓','B公寓','C農場','D別墅'] #temp reset
list2 = [100,200,300,400]
list2.extend(list1)
print('變身後:', list2)
```

```python
#合併(方法2)
list1 = ['A大樓','B公寓','C農場','D別墅'] #temp reset
list2 = [100,200,300,400]

print('變身前:', list1, list2)
x = list1 + list2
print('變身後:', x)
```

### Data & Time
> https://docs.python.org/3/library/datetime.html

#### Time
* Python: datetime
* Numpy: datetime64(faster)

#### Time Index
* DatetimeIndex (Timestamp / time point)
* PeriodIndex (Period of time - with start time)
* TimedeltaIndex(Time difference - no start time)

#### Python

```python
#Python
import datetime as dt


now = dt.datetime.now() #時分秒
today = dt.date.today() #今天
print("時分秒: ", now)
print("今天: ", today)
print("-" * 20)

print(now.year)
print(now.month)
print(now.day)
print(now.weekday())
print(now.ctime())
print("-" * 20)

now = dt.date.today()
print("now= ", now)
print("now hh24= ", dt.datetime.now().hour)
print("-" * 20)

#[Format]
print("-" * 20, "[Format]", "-" * 20)
print("~" * 60)
print("<strftime: format date object>")
print(now.strftime("%b %d %Y %H:%M:%S"))
print(now.strftime("%Y%m%d"))
print("~" * 60)
#String to Date
StrToDate = "20220403 13:14:20"
x = dt.datetime.strptime(StrToDate, "%Y%m%d %H:%M:%S")
print(x)
print(type(x))
print("~" * 60)
print("<strptime: parse time string with a time format>")
x = dt.datetime.strptime('Mon Sep 28 2020', "%a %b %d %Y")
print("x= ", x , "type=", type(x))
print("~" * 60, "\n")

#yyyymmdd = ymd.format(now.year, now.month, now.day)
ymd = "{}{}{}"
yyyymmdd = ymd.format(now.strftime("%Y"), now.strftime("%m"), now.strftime("%d"))
print(yyyymmdd)
print("~" * 20, "\n")

#print(dt.datetime.now().isoformat(timespec="microseconds"))

print("today day is: ", now.day)
print(now.day-1)
print("~" * 20, "\n")


#String to Date
```

##### 加減日子

```python
x = now + dt.timedelta(days=-10)
print("now=", now)
print("day-10= ", x)

x = now + dt.timedelta(hours=30)
print("now=", now)
print("hours+30= ", x)
```

```python
import dateutil.relativedelta as dtre
dt_str = dt.date.today() + dtre.relativedelta(days=-1)
print(dt_str.strftime("%Y%m%d"))
dt_str = dt.date.today() + dtre.relativedelta(months=-1)
print(dt_str.strftime("%Y%m%d"))
dt_str = dt.date.today() + dtre.relativedelta(months=-1, days=-1)
print(dt_str.strftime("%Y%m%d"))
```

### 物件導向

```python
class Check_Order:

    @classmethod
    def to_string(str, format):
        order, account, mobile = format.split(",")
        return str(order, account, mobile)

    def __init__(self, order_id, account_no, mobile_no):
        self.order = order_id
        self.account = account_no
        self.mobile = mobile_no

    #show account details
    def show_ac_mob(self):
        self.account = 12345
        return f"Account: {self.account}, Mobile: {self.mobile}"

    #show like what
    def show_like(self, something):
        return f"he like {something} ..."
```

```python
x = Check_Order("34","523452","2342342").show_ac_mob()
print(x)
```

```python
x = Check_Order("34","523452","2342342").show_like("football")
print(x)
```

```python
x = Check_Order.to_string("34, 523452, 2342342").show_ac_mob()
print(x)
```
